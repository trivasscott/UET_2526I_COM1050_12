```python
#W8A1
class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h
    
    def sa(self):
        return f"{6.28 * self.r * self.r + 6.28 * self.r * self.h:.2f}"
    
    def v(self):
        return f"{3.14 * self.r * self.r * self.h:.2f}"

r, h = map(float, input().split())
c = Cylinder(r, h)
print(c.sa(), c.v())
```

```python
#W8A2
class Date:
    cnt = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    def __init__(self, d, m, y):
        self.d = d
        self.m = m
        self.y = y
    
    def nxt(self):
        k = self.cnt[self.m - 1] + (self.m == 2 and (self.y % 400 == 0 or (self.y % 4 == 0 and self.y % 100 != 0)))
        if 1 <= self.d <= k:
            nxt_d = self.d + 1
            nxt_m = self.m
            nxt_y = self.y
            if nxt_d > k:
                nxt_d = 1
                nxt_m += 1
            if nxt_m > 12:
                nxt_m = 1
                nxt_y += 1
            print(nxt_d, nxt_m, nxt_y, sep = '/')
        else:
            print("INVALID")

d, m, y = map(int, input().split('/'))
D = Date(d, m, y)
D.nxt()
```

```python
#W8A3
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def is_origin(self):
        return self.x == self.y == 0
    
    def on_yaxis(self):
        return self.x == 0

    def on_xaxis(self):
        return self.y == 0

    def dist_to_o(self):
        return round((self.x * self.x + self.y * self.y) ** .5, 2)

x, y = map(float, input().split())
M = Point(x, y)
print(M.is_origin(), M.on_xaxis(), M.on_yaxis(), M.dist_to_o())
```

```python
#W8A5
class ShoppingCart:
    def __init__(self):
        self.dct = {}
    
    def add_item(self, name, price):
        if name not in self.dct:
            self.dct[name] = int(price)
        else:
            self.dct[name] += int(price)
        print("Added.")
    
    def del_item(self, name):
        if name in self.dct:
            self.dct.pop(name)
            print("Deleted.")
        else:
            print("Not found.")
    
    def is_empty(self):
        print("Not empty" if len(self.dct) > 0 else "Empty.")

    def calc(self):
        print(f"Total value: {sum(self.dct.values())}.")
    
    def view(self):
        print("Your cart:")
        print(self.dct)

    def clear(self):
        self.dct = {}
        print("Cart is now empty.")

sc = ShoppingCart()

while True:
    opt = input('''
    Thao tac?
    2. add item
    3. delete item
    4. check 4 emptiness
    5. calc total value
    6. view cart
    7. empty cart
    ''')
    
    if opt == '2':
        name, price = input('Item name & price?').split()
        sc.add_item(name, price)
    elif opt == '3':
        name = input('Item?')
        sc.del_item(name)
    elif opt == '4':
        sc.is_empty()
    elif opt == '5':
        sc.calc()
    elif opt == '6':
        sc.view()
    elif opt == '7':
        sc.clear()
    else:
        print("Invalid!")

    if input('Thoat?').lower() == 'yes':
        break
```

```python
#W8A6
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h
    
    def area(self):
        return f"{self.w * self.h:.2f}"
    
    def perimeter(self):
        return f"{(self.w + self.h) * 2:.2f}"

    def scale(self, k):
        self.w *= k
        self.h *= k

w, h, k = map(float, input().split())
r = Rectangle(w, h)

# print(r.area())
# print(r.perimeter())
r.scale(k)
print(r.area(), r.perimeter())
```

```python
#W8A7
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

class Fraction:

    def __init__(self, num, den):
        x = gcd(num, den)
        self.num = num // x
        self.den = den // x
    
    def __add__(self, other):
        return Fraction(self.num * other.den + other.num * self.den, self.den * other.den)
    
    def __sub__(self, other):
        return Fraction(self.num * other.den - other.num * self.den, self.den * other.den)

    def __mul__(self, other):
        return Fraction(self.num * other.num, self.den * other.den)
    
    def __truediv__(self, other):
        return Fraction(self.num * other.den, self.den * other.num)

    def __str__(self):
        return f"{self.num}/{self.den}"
    
a, b, op, c, d = input().split()
a, b, c, d = int(a), int(b), int(c), int(d)

f1, f2 = Fraction(a, b), Fraction(c, d)

if op == '+':
    print(f1 + f2)
elif op == '-':
    print(f1 - f2)
elif op == '*':
    print(f1 * f2)
else:
    print(f1 / f2)
```

```python
#W8A8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance
    
    def deposit(self, x):
        self.balance += x

    def withdraw(self, x):
        if self.balance >= x:
            self.balance -= x
    
    def check(self):
        return f"{self.balance:.2f}"

    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"

owner, initial_balance = input().split()
initial_balance = int(initial_balance)

ba = BankAccount(owner, initial_balance)

for _ in range(int(input())):
    call, x = input().split()
    x = float(x)
    if call == "DEPOSIT":
        ba.deposit(x)
    else:
        ba.withdraw(x)

print(ba.check())
```

```python
#W8A9
class Student:
    def __init__(self, name):
        self.name = name
        self.total = 0
        self.cnt = 0
    
    def add_score(self, score):
        self.total += score
        self.cnt += 1
    
    def avg(self):
        return self.total / max(1, self.cnt)

    def rank(self):
        avg = self.avg()
        return "Excellent" if avg >= 8 else "Good" if avg >= 6.5 else "Average" if avg >= 5 else "Poor"

    def __str__(self):
        return f"{self.name} {self.avg():.2f} {self.rank()}"

name = input()

s = Student(name)

for _ in range(int(input())):
    sb, sc = input().split()
    s.add_score(float(sc))

print(s)
```

```python
#W8A10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year
    
class Library:
    def __init__(self):
        self.dct_1 = {}
        self.dct_2 = {}

    def add(self, book):
        author = book.author
        year = book.year
        if author in self.dct_1:
            self.dct_1[author] += 1
        else:
            self.dct_1[author] = 1
        if year in self.dct_2:
            self.dct_2[year] += 1
        else:
            self.dct_2[year] = 1
    
    def count_by_author(self, name):
        return self.dct_1[name] if name in self.dct_1 else 0

    def find_by_year(self, year):
        return self.dct_2[year] if year in self.dct_2 else 0

l = Library()

for _ in range(int(input())):
    call, x = input().split()
    if call == 'ADD':
        title, author, year = x.split(';')
        b = Book(title, author, year)
        l.add(b)
    elif call == 'COUNT':
        print(l.count_by_author(x))
    else:
        print(l.find_by_year(x))
```

```python
#W8A11
import math

class Point:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

class Vector(Point):
    def __init__(self, a, b, c = None):
        if c is None:
            self.x = b.x - a.x
            self.y = b.y - a.y
            self.z = b.z - a.z
        else:
            super().__init__(a, b, c)

    def cross(self, other):
        x = self.y * other.z - self.z * other.y
        y = self.z * other.x - self.x * other.z
        z = self.x * other.y - self.y * other.x
        return Vector(x, y, z)
    
    def dot(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z
    
    def len(self):
        return math.sqrt(self.x * self.x + self.y * self.y + self.z * self.z)
    
class Plane:
    def __init__(self, a, b, c):
        ab = Vector(a, b)
        ac = Vector(a, c)
        self.normal = ab.cross(ac)
    
    def phi(self, other):
        num = abs(self.normal.dot(other.normal))
        den = self.normal.len() * other.normal.len()
        return f"{math.degrees(math.acos(num / den)):.2f}"

x, y, z = map(float, input().split())
a = Point(x, y, z)
x, y, z = map(float, input().split())
b = Point(x, y, z)
x, y, z = map(float, input().split())
c = Point(x, y, z)
x, y, z = map(float, input().split())
d = Point(x, y, z)

print(Plane(a, b, c).phi(Plane(b, c, d)))
```

```python
#W8A12
class ComplexNumber:
    def __init__(self, re, im):
        self.re = re
        self.im = im
    
    def __mul__(self, other):
        a, b, c, d = self.re, self.im, other.re, other.im
        return ComplexNumber(a * c - b * d, a * d + b * c)

    def __str__(self):
        return f"{self.re} {self.im}"

a, b = map(float, input().split())
x = ComplexNumber(a, b)
c, d = map(float, input().split())
y = ComplexNumber(c, d)

print(x * y)
```

```python
#W8A13
class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    
    def getArea(self):
        a, b, c = self.a, self.b, self.c
        p = a + b + c
        if 2 * max(a, b, c) < p:
            return f"{(p * (p - 2 * a) * (p - 2 * b) * (p - 2 * c)) ** .5 / 4:.2f}"
        else:
            return "invalid"

a, b, c = map(float, input().split())
print(Triangle(a, b, c).getArea())
```
