```python
#SumPrimesInRange
a, b = map(int, input().split())
s, i = [0] * 2 + [1] * b, 2
while i * i <= b:
    if s[i]:
        j = i * i
        while j <= b:
            s[j] = 0
            j += i
    i += 1
print(sum(k * s[k] for k in range(a, b + 1)))
```

```python
#LargestPrimeFactor
n = int(input())
ans = 2
for i in range(2, int(n ** 0.5) + 3):
    while n % i == 0:
        ans = i
        n //= i
if n > 2:
    ans = n
print(ans)
```

```python
#PerfectSquaresDistinctDigits
n = int(input())
def dif(x):
    l = [0] * 10
    while x:
        l[x % 10] += 1
        x //= 10
    return max(l) <= 1

if n:
    for i in range(1, int(n ** 0.5) + 1):
        if dif(i * i):
            print(i * i, end=' ')
else:
    print('No number')
```

```python
#LongestCollatzUpToN
n = int(input())
f = {1: 1}
def clen(x: int) -> int:
    if x in f:
        return f[x]
    if x % 2 == 0:
        f[x] = clen(x // 2) + 1
    else:
        f[x] = clen(3 * x + 1) + 1
    return f[x]
id, ans = 1, 1
for i in range(2, n + 1):
    if clen(i) > ans:
        id, ans = i, f[i]
print(id, ans)
```

```python
#EvenDivisors
n = int(input())
i, ans = 1, 0
while i * i <= n:
    if n % i == 0:
        ans += (i % 2 == 0) + (n // i != i and n // i % 2 == 0)
    i += 1
print(ans)
```
