```python
#W6A4
s = input().split()
d = {}
for i in s:
    a = b = ''
    for j in i:
        if j == ':' and not b:
            a, b = b, a
        else:
            a += j
    if b in d:
        d[b] += [a]
    else:
        d[b] = [a]
print(d)
```

```python
#W6A9
d = {}
for i in range(2):
    s = input()[1:-1]
    if s:
        s = s.split(',')
    for j in s:
        a, b = j.split(':')
        a = a.strip().strip("'")
        d[a] = d[a] + int(b) if a in d else int(b)
print({i: d[i] for i in sorted(d.keys())})
```

```python
#W6A10
a = list(map(int, input().split()))
k = int(input())
ans = []
for i in range(1, len(a)):
    if k - a[i] in a[:i]:
        ans += [k - a[i]]
print(list((i, k - i) for i in sorted(set(ans))))
```
