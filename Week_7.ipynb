```python
#W7A1
def binary_search(a: list[int], l: int, r: int, b: int) -> int:
    l, r = 0, len(a) - 1
    while l <= r:
        mid = l + r >> 1
        if a[mid] >= b:
            r = mid - 1
        else:
            l = mid + 1
    return l if l < len(a) and a[l] == b else -1
a, b = list(map(int, input().split())), int(input())
print(binary_search(a, 0, len(a) - 1, b))
```

```python
#W7A3
def bubble_sort(a: list[int]) -> list[int]:
    for i in range(len(a)):
        for j in range(i + 1, len(a)):
            if a[i] > a[j]:
                a[i], a[j] = a[j], a[i]
    return a

def selection_sort(a: list[int]) -> list[int]:
    for i in range(len(a)):
        id = i
        for j in range(i + 1, len(a)):
            if a[j] < a[id]:
                id = j
        a[i], a[id] = a[id], a[i]
    return a

a = list(map(int, input().split()))
assert bubble_sort(a) == selection_sort(a) == sorted(a)
print(*bubble_sort(a))
```

```python
#W7A5
a = sorted(list(map(int, input().split())))
x = int(input())

lf, rt = [0] * len(a), [0] * len(a)

for i in range(len(a)):
    lf[i] = lf[i - 1] if i and a[i] == a[i - 1] else i
for i in range(len(a) - 1, -1, -1):
    rt[i] = rt[i + 1] if i + 1 < len(a) and a[i] == a[i + 1] else i

l, r, ans = 0, len(a) - 1, 0

while l < r:
    s = a[l] + a[r]
    if s == x:
        ans += (rt[l] - lf[l] + 1) * (rt[r] - lf[r] + 1)
        l = rt[l] + 1
        r = lf[r] - 1
    elif s < x:
        l = rt[l] + 1
    else:
        r = lf[r] - 1

print(ans)
```

```python
#W7A7
a = list(map(int, input().strip()[1:-1].split(',')))
n = len(a)
tail = [min(a) - 67] + [max(a) + 67] * n
ans = 0
for i in range(n):
    l, r = 0, i
    while l <= r:
        mid = l + r >> 1
        if tail[mid] < a[i]:
            l = mid + 1
        else:
            r = mid - 1
    if r >= 0 and tail[r] < a[i]:
        ans = max(ans, r + 1)
        tail[r + 1] = min(tail[r + 1], a[i])
print(ans)
```

```python
#W7A8

a = list(list(int(j) for j in i.split(',')) for i in ''.join(input().strip()[2:-2].split()).split('],['))
a = sorted(list([i[1] - i[0] + 1, i[0], i[1]] for i in a))

q = list(map(int, input().strip()[1:-1].split(',')))
q = sorted(list([q[i], i] for i in range(len(q))))

n = len(q)
ans = [-1] * n

def bin_l(x: int) -> int:
    l, r = 0, n - 1
    while l <= r:
        mid = l + r >> 1
        if q[mid][0] >= x:
            r = mid - 1
        else:
            l = mid + 1
    return l if l < n else -1

def bin_r(x: int) -> int:
    l, r = 0, n - 1
    while l <= r:
        mid = l + r >> 1
        if q[mid][0] <= x:
            l = mid + 1
        else:
            r = mid - 1
    return r if r >= 0 else -1

par = [i for i in range(n + 1)]

def find(i: int) -> int:
    if par[i] == i:
        return i
    par[i] = find(par[i])
    return par[i]

for [len, i, j] in a:
    if i > q[-1][0] or j < q[0][0]:
        continue
    l = bin_l(i)
    r = bin_r(j)
    assert l != -1 and r != -1
    cur = find(l)
    while cur <= r:
        ans[q[cur][1]] = len
        par[cur] = find(cur + 1)
        cur = find(cur)

print(ans)
```

```python
#W7A9

a = list(map(int, input().strip()[1:-1].split(',')))

n = len(a)
ans = [0] * n

b = [[a[i], i] for i in range(n)]

st = [[0, n - 1, 0]]

while st:
    l, r, fl = st.pop()
    if l >= r:
        continue
    mid = l + r >> 1
    if fl:
        i, j = l, mid + 1
        tmp = []
        while i <= mid and j <= r:
            if b[j][0] < b[i][0]:
                tmp += [b[j]]
                j += 1
            else:
                ans[b[i][1]] += j - mid - 1
                tmp += [b[i]]
                i += 1
        while i <= mid:
            ans[b[i][1]] += j - mid - 1
            tmp += [b[i]]
            i += 1
        while j <= r:
            tmp += [b[j]]
            j += 1
        for i in range(r - l + 1):
            b[l + i] = tmp[i]
    else:
        st += [[l, r, 1], [l, mid, 0], [mid + 1, r, 0]]

print(ans)
```

```python
#W7A10

a = list(map(int, input().strip()[1:-1].split(',')))
m = int(input())

n = len(a)
pre = [0] * n
for i in range(n):
    pre[i] = ((pre[i - 1] + a[i] % m) + m) % m if i else ((a[i] % m) + m) % m

b = [[pre[i], i + 1] for i in range(n)]

st = [[0, n - 1, 0]]

ans = max(pre)

while st:
    l, r, fl = st.pop()
    if l >= r:
        continue
    mid = l + r >> 1
    if fl:
        i, j = l, mid + 1
        tmp = []
        while i <= mid and j <= r:
            if b[i][0] <= b[j][0]:
                tmp += [b[i]]
                i += 1
            else:
                ans = max(ans, b[j][0] - b[i][0] + m)
                tmp += [b[j]]
                j += 1
        while i <= mid:
            tmp += [b[i]]
            i += 1
        while j <= r:
            tmp += [b[j]]
            j += 1
        for i in range(r - l + 1):
            b[l + i] = tmp[i]
    else:
        st += [[l, r, 1], [l, mid, 0], [mid + 1, r, 0]]
print(ans)
```
