```python
#W9A1
a = list(map(int, input().split()))
b = list(map(int, input().split()))
n, m = len(a), len(b)
i, j = 0, 0
ans = []
while i < n and j < m:
    if a[i] < b[j]:
        ans += [a[i]]
        i += 1
    else:
        ans += [b[j]]
        j += 1
while i < n:
    ans += [a[i]]
    i += 1
while j < m:
    ans += [b[j]]
    j += 1
print(*ans)
```

```python
#W9A2
l = []
while 1:
    try:
        a = list(map(int, input().split()))
        l += [a]
    except:
        break
v = l[-1][0]
l.pop()
m, n = len(l), len(l[0])
h, w = min(m, n), max(m, n)
for i in range(m):
    for j in range(n):
        s = i + j
        if s < h:
            c = s * (s + 1) // 2
        elif s < w:
            c = s * h - h * (h - 1) // 2
        else:
            r = m + n - 1 - s
            c = m * n - r * (r + 1) // 2
        b = i - max(0, s - n + 1)
        if v == c + b + 1:
            print(l[i][j])
            break
```

```python
#W9A3
s = input()
l = []
for i in s:
    l += [i]
idx = 0
for i in range(1, len(l)):
    if l[i] <= l[idx]:
        idx = i
if idx:
    l[0], l[idx] = l[idx], l[0]
else:
    l[-1], l[-2] = l[-2], l[-1]
print(''.join(l[l[0] == '0':]))
```

```python
#W9A4
T = int(input())
for _ in range(T):
    s, t = input().strip(), input().strip()
    m, n, j, yes = len(s), len(t), 0, 1
    for i in t:
        while j < m and s[j].upper() != i:
            j += 1
        yes  &= (j != m)
    print("YES" if yes else "NO")
```

```python
#W9A5
s, t = list(map(int, input().split())), list(map(int, input().split()))
m, n = len(s), len(t)
s = s + [min(s) - 67] + t
pi = [0]
yes = 0
for i in range(1, m + n + 1):
    j = pi[i - 1]
    while j and s[i] != s[j]:
        j = pi[j - 1]
    if s[i] == s[j]:
        j += 1
    if j == m:
        yes = 1
        break
    pi += [j]
print("YES" if yes else "NO")
```

```python
#W9A6
n = int(input())
a = list(map(int, input().split()))
i = n - 2
while i >= 0 and a[i] > a[i + 1]:
    i -= 1
if i != -1:
    j = n - 1
    while a[j] <= a[i]:
        j -= 1
    a[i], a[j] = a[j], a[i]
    a[i + 1:] = a[i + 1:][::-1]
else:
    a.reverse()
print(*a)
```

```python
#W9A7
a = [1, 0] + list(map(int, input().split())) + [0, 1]
k = int(input())
pre, cnt = 0, 0
for i in range(1, len(a)):
    if a[i]:
        cnt += (i - pre) // 2 - 1
        pre = i
print(cnt >= k)
```

```python
#W9A8
s = input().strip()
lst = [(0, 0)]
for i in s:
    lst += [(lst[-1][0] + (1 if i == 'R' else -1 if i == 'L' else 0), lst[-1][1] + (1 if i == 'U' else -1 if i == 'D' else 0))]
print(len(set(lst)) != len(lst))
```

```python
#W9A9
a, b, c, d, e = map(int, input().split())
if a > b:
    a, b = b, a
if c > d:
    c, d = d, c
if b > d:
    a, b, c, d = c, d, a, b
if c > e:
    c, e = e, c
if b < e:
    print(max(b, c))
else:
    print(max(a, e))
```

```python
#W9A10
a = sorted(set(list(map(int, input().split()))))
b = sorted(set(list(map(int, input().split()))))
n, j, ans = len(b), 0, 0
for i in a:
    while j + 1 < n and abs(b[j + 1] - i) <= abs(b[j] - i):
        j += 1
    ans = max(ans, abs(b[j] - i))
print(ans)
```
