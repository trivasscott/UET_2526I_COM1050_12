```python
#W10A3
def isHappy(n: int) -> bool:
    lst = set()
    tmp = n
    while tmp != 1:
        sum = 0
        while tmp:
            x = tmp % 10
            sum += x * x
            tmp //= 10
        if sum == 1:
            break
        if sum in lst:
            return 0
        lst.add(sum)
        tmp = sum
    return 1

try:
    print("YES" if isHappy(int(input())) else "NO")
except ValueError:
    print('ngu')
```
